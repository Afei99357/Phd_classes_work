```python
# Create a plot showing 1000 points on a sine wave
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create four sine waves of different frequencies, offset vertically by 1
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.sin(2*x) + 3
y3 = np.sin(3*x) + 6
y4 = np.sin(4*x) + 9

# Plot all four sine waves
for p in [y1, y2, y3, y4]:
    # remove the ticks anbd labels from the plot
    plt.plot(x, p)
    plt.xticks([])
    plt.yticks([])

# In another figure on the right, create a horizontal bar chart showing amplitudes
plt.figure()
# Show each bar in a different color to match the plot above
plt.barh([1, 2, 3, 4], [1, 4, 2, 5],
            color=['b', 'orange', 'g', 'r'])

# Remove the x and y labels and ticks
plt.xticks([])
plt.yticks([])
plt.show()

# now, save it to a file
plt.savefig('my_figure.png')
```
